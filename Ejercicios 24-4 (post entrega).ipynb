```python
Nombres=[]
Nombresmenores=[]
for i in range(int(input("Ingrese el número de nombres a ingresar: "))):
    a=input(("Ingrese un nombre "))
    if a.lower() in Nombresmenores:
        a=input(("Este nombre ya ha sido ingresado, favor ingresar otro: "))
    Nombres.append(a)
    Nombresmenores.append(a.lower())
    print("\n\n")
for i in range(len(Nombres)):
    if (i+1)%3==0:
        print("Buenas noches, Sr./Sra. " +Nombres[i])
    elif (i+1)%2==0:
        print("Buenas tardes, Sr./Sra, " +Nombres[i])
    else:
        print("Buenos días, Sr./Sra. " +Nombres[i])
```

```python
Invitados=[]
maxi=(int(input("Ingrese el máximo de invitados\n")))
if maxi<1:
    max=(int(input("Imposible, ingrese de nuevo\n")))
x=int(input("Ingrese el tamaño de la lista inicial\n"))

while 1:
    if x>maxi:
        x=int(input("El tamaño de esta lista es demasiado grande, ingrese de nuevo\n"))
    elif x<0:
        x=int(input("Imposible, ingrese de nuevo\n"))
    else:
        break
        
if x==0:
    print("Usted no tiene invitados")
else:
    for i in range(x):
            Invitados.append(input("Ingrese un nombre\n"))
print("\nEstas son las acciones que usted puede hacer: \nAgregar invitado\nEliminar invitado\nCambiar invitado\nAyuda\nFinalizar\n")
while 1:
    Inst=input("¿Qué desea hacer?\n")
    if Inst.lower()=="finalizar":
        print("Finalizando\n")
        break
    elif Inst.lower()=="agregar invitado":
        if len(Invitados)<maxi:
            a=(input("Ingrese el nombre de su invitado\n"))
            if a in Invitados:
                print(str(a)+" ya se encuentra en la lista")
            else:
                Invitados.append(a)
                print("Se ha agregado a "+str(Invitados[len(Invitados)-1])+" a su lista")
        else:
            print("La lista está llena")
    elif Inst.lower()=="eliminar invitado":
        a=input("Ingrese el nombre del rechazado\n")
        if a in Invitados:
            Invitados.remove(a)
            print(str(a)+" ha sido eliminado")
        else:
            print(str(a)+" no se encuentra en la lista")
    elif Inst.lower()=="cambiar invitado":
        a=input("¿A quién quiere cambiar?\n")
        if a in Invitados:
            b=input("¿A quién quiere poner en su lugar?")
            if b in Invitados:
                print("Este invitado ya se encuentra en la lista, no hace falta cambiarlo por "+str(a))
            else:
                Invitados[Invitados.index(a)]=b
        else:
            print(str(a)+ " no se encuentra en la lista")
    elif Inst.lower()=="ayuda":
        print("Estas son las acciones que usted puede hacer: \nAgregar invitado\nEliminar invitado\nCambiar invitado\nAyuda\nFinalizar\n")
    else:
        print("Comando inaceptable")
if len(Invitados)>0:
    print("Imprimiendo lista de "+str(len(Invitados))+" invitados\n")
    Invitados.sort()
    for i in range(len(Invitados)):
        print(Invitados[i])
else:
    print("Usted no tiene invitados")
```

```python
lista=[]
largo=int(input("Ingrese el largo de la lista\n"))
for i in range(largo):
    a=int(input("Ingrese un número par\n"))
    while (a%2)!=0:
        a=int(input("Este no es un número par\nIngrese un número par\n"))
    lista.append(a)
print("Imprimiendo lista original")
print(lista)
print("Realizando operaciones, imprimiendo conversión número por número")
for i in range(largo):
    lista[i]=lista[i]**3
    print(lista)
```

```python
lista=[]
listalower=[]
long=int(input("Ingrese el largo de la lista\n"))
while long<1:
    long=int(input("Número inválido, vuelva a ingresar el largo de la lista\n"))
while len(lista) in range(long):
    name=input("Ingrese un nombre para agregar a la lista\n")
    if name.lower()=="admin":
        print("No puede ingresar un administrador a la lista\n")
    elif name.lower() in listalower:
        print("Este nombre ya se encuentra en la lista\n")
    else:
        lista.append(name)
        listalower.append(name.lower())
        print("Agregado")
print(lista)
```

```python
diccionario={"NOMBRE": input("Ingrese su nombre\n"), "APELLIDO": input("Ingrese su apellido\n")}
extras=int(input("¿Cuántos ámbitos más desea ingresar?"))
while extras<0:
    extras=int(input("No puede ingresar un número negativo para esta pregunta\n¿Cuántos ámbitos más desea ingresar?"))
for i in range(extras):
    a=input("Ingrese el ámbito a ingresar (Ejemplo: Dirección, edad)\n")
    while a.upper() in diccionario:
        a=input("Usted ya ha ingresado este ámbito, ingrese uno nuevo")
    b=input("Ingrese la respuesta para el ámbito a ingresar (Ejemplo: Calle sin nombre 123, 18)\n")
    diccionario[a.upper()]=b
print(diccionario)
```
